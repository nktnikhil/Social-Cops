### Detecting Seasonality and its decomposition

```python
import statsmodels.api as sm
import math
math.tan(5*3.14/(180))
t = 0;
if len(p) < 4:
    print('This commodity doesn not have enough data for time series ' )
else :
    t = 1;
    p = pd.Series(p)
    p = p.interpolate()
    p = pd.Series(p.dropna().values)
    res = sm.tsa.seasonal_decompose(p,freq=4,model='additive')
    res1 = sm.tsa.seasonal_decompose(p,freq=4,model='multiplicative')
    k1 = np.zeros(len(res.seasonal));
    k2 = np.zeros(len(res1.seasonal));
    for i in range(len(res.seasonal)):
        k1[i] = i;
        k2[i] = i;
    ind = np.argpartition(res.seasonal, -4)[-4:].values
    
   
    ind1 = np.argpartition(res1.seasonal, -4)[-4:].values
    
    z = np.polyfit(ind,res.seasonal[ind],1)
    z1 = np.polyfit(ind1,res1.seasonal[ind1],1)
    
    if abs(z[0]) < abs(z1[0]):
        resplot = res.plot()
        resplot.savefig('Seasonal_Decompose.png', dpi=150)
        print('This series has additive seasonality');
        series = ['Additive'];
    else :
        print('This series has multiplicative seasonality');
        series = ['Multiplicative'];
        resplot = res1.plot()
        resplot.savefig('Seasonal_Decompose.png', dpi=150)
        res = res1
    
```

```python
kharif_month = ['July','August','September','October'];
rabi_month = ['November','December','January','February','March'];
```

## Flagging highest price fluctuation of commodities in seasons and year

```python
for c1 in range(sort_df['Year'].values[0],sort_df['Year'].values[len(sort_df['Year']) - 1]+1,1):
    b1 = sort_df.loc[sort_df['Year'] == c1];
    community = set(b1['Commodity'])
    y1 = np.zeros(len(community))
    kharif = np.zeros(len(y1))
    
    rabi = np.zeros(len(y1))
    s1 = 0;
    for da in community:
        ba = b1.loc[b1['Commodity'] == da]
        
        kha  = 0;
        ra = 0;
        for ma in kharif_month:
            if ba.loc[ba['Month'] == ma].shape[0] > 0:
                kha = kha+1;
        for ma in rabi_month:
            if ba.loc[ba['Month'] == ma].shape[0] > 0:
                ra = ra+1;        
        kharif_season_price = np.zeros(kha);
        
        rabi_season_price = np.zeros(ra);
        whole_year_price = np.zeros(ba.shape[0]);
        if kha> 0:
            i = 0;
            for ma in kharif_month:
                if ba.loc[ba['Month'] == ma].shape[0] > 0:
                    kharif_season_price[i] = ba.loc[ba['Month'] == ma]['modal_price'].values ;
                    i = i+1;
        if ra > 0 :           
            i = 0;
            for ma in rabi_month:
                if ba.loc[ba['Month'] == ma].shape[0] > 0:
                    rabi_season_price[i] = ba.loc[ba['Month'] == ma]['modal_price'].values ;
                    i = i+1;
        i = 0;
        for ma in range(len(whole_year_price)):
            whole_year_price[i] = ba['modal_price'].values[i];
            i = i+1;
        y1[s1] = np.var(whole_year_price);
        if kha>0:
            kharif[s1] = np.var(kharif_season_price);
        if ra > 0:    
            rabi[s1] = np.var(rabi_season_price);
        s1 = s1+1;
    g1 = list(community)  
    
    print("In year ",c1,",commodity with highest price fluctuaton:-",g1[np.argmax(y1)]);
    if sum(kharif) > 0:
        print("In kharif season of year",c1,"commodity with highest price fluctuation:-",g1[np.argmax(kharif)]);
    if sum(rabi) > 0:
        print("In rabi season of year",c1,"commodity with highest price fluctuation:-",g1[np.argmax(rabi)]);
    print(" ")
```
